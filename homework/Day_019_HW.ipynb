```python
# 載入需要的套件
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns # 另一個繪圖-樣式套件

# 忽略警告訊息
%matplotlib inline
plt.style.use('ggplot')
import warnings
warnings.filterwarnings('ignore')

# 設定 data_path
dir_data = '/content/drive/My Drive/Colab Notebooks'
```

```python
# 讀取檔案
f_app = os.path.join(dir_data, 'application_train.csv')
print('Path of read in data: %s' % (f_app))
app_train = pd.read_csv(f_app)
app_train.head()
```

```python
print(app_train['HOUSETYPE_MODE'])
print('---')
print(app_train['HOUSETYPE_MODE'].value_counts()[0:4])
print('---')
print(app_train['HOUSETYPE_MODE'].describe())
print('---')
print("That is", len(app_train['HOUSETYPE_MODE'].unique()))
print('---')
print('The member is',app_train['HOUSETYPE_MODE'].unique()[1])
```

```python
# 使用不同的 HOUSETYPE_MODE 類別繪製圖形, 並使用 subplot 排版
"""
Your Code Here
"""

unique_house_type = app_train['HOUSETYPE_MODE'].unique()
print(unique_house_type)
nrows = len(unique_house_type)
ncols = nrows // 2

plt.figure(figsize=(10,30))
for i in range(len(unique_house_type)):
    plt.subplot(nrows, ncols, i+1)
    # plt.subplot（j,k,i）表示將整個圖像窗口分成j列k行，當前位置為i。

    
    """
    Your Code Here
    """
    app_train.loc[app_train['HOUSETYPE_MODE'] == unique_house_type[i], 'AMT_CREDIT']
    app_train.loc[app_train['HOUSETYPE_MODE'] == unique_house_type[i], 'AMT_CREDIT'].hist()
    
    
    plt.title(str(unique_house_type[i]))
plt.show()    
```
